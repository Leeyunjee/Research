```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

## **US - New York Times**

**New York Times (2014 ~ 2017)**

```python
from datetime import datetime, timedelta

df = []

start_date = datetime(2014, 1, 1)
end_date = datetime(2017, 12, 31)
date_list = []


while start_date <= end_date:
  date_list.append(start_date.strftime("%Y/%m/%d"))
  start_date += timedelta(days=1)

for d in date_list:
  url = f"https://archive.nytimes.com/www.nytimes.com/indexes/{d}/todayspaper/index.html"

  response = requests.get(url)
  html = response.text

  soup = BeautifulSoup(html, 'html.parser')

  dd = datetime.strptime(d, "%Y/%m/%d")
  newd = dd.strftime("%Y-%m-%d")

  tit_texts = soup.find('div', class_='aColumn')

  for tit_text in tit_texts.find_all('a'):
    df.append([newd, tit_text.get_text(strip=True)])

df = pd.DataFrame(df, columns=["Date", "Headline"])
df = df[df["Headline"]!=""].reset_index(drop=True)
```

```python
df = df.groupby("Date")["Headline"].agg(" ".join).reset_index()
df
```

**New York Times (2018 ~ 2022)**

```python
from datetime import datetime, timedelta

df = []

start_date = datetime(2018, 1, 1)
end_date = datetime(2022, 12, 31)
date_list = []


while start_date <= end_date:
  date_list.append(start_date.strftime("%Y/%m/%d"))
  start_date += timedelta(days=1)

for d in date_list:
  url = f"https://www.nytimes.com/issue/todayspaper/{d}/todays-new-york-times"

  response = requests.get(url)
  html = response.text

  soup = BeautifulSoup(html, 'html.parser')

  dd = datetime.strptime(d, "%Y/%m/%d")
  newd = dd.strftime("%Y-%m-%d")

  tit_texts = soup.find_all('h3', class_='css-miszbp e1hr934v2')

  for tit_text in tit_texts:
    df.append([newd, tit_text.get_text(strip=True)])

  tit_texts2 = soup.find_all('h2', class_='css-ds6ff4 e1b0gigc0')

  for tit_text2 in tit_texts2[:10]:
    df.append([newd, tit_text2.get_text(strip=True)])

df2 = pd.DataFrame(df, columns=["Date", "Headline"])
df2 = df[df["Headline"]!=""].reset_index(drop=True)
```

```python
df2 = df.groupby("Date")["Headline"].agg(" ".join).reset_index()
df2
```

```python
news = pd.concat([df, df2], ignore_index=True)
news
```

```python
news.to_csv("US_news.csv")
```

## **UK - The Independent**

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

from datetime import datetime, timedelta
import re

df = []

start_date = datetime(2014, 1, 1)
end_date = datetime(2022, 12, 31)
date_list = []


while start_date <= end_date:
  date_list.append(start_date.strftime("%Y-%m-%d"))
  start_date += timedelta(days=1)

headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"}

for d in date_list:
  url = f"https://www.independent.co.uk/archive/{d}"

  response = requests.get(url, headers=headers)
  html = response.text

  soup = BeautifulSoup(html, 'html.parser')

  tit_texts = soup.find('div', class_=re.compile("sc-9rjm6m-0"))
  if tit_texts is not None:
    tit_text = tit_texts.find_all('a')
    for tit in tit_text[1:21]:
      df.append([d, tit.text])

df3 = pd.DataFrame(df, columns=["Date", "Headline"])
df3
```

```python
df3 = df3.groupby("Date")["Headline"].agg(" ".join).reset_index()
df3
```

```python
df3.to_csv("UK_news.csv")
```
